# Brazil


```
import ipeadatapy
import pandas as pd

# Dicion√°rio com o nome da s√©rie e seu c√≥digo no IPEA
series = {
    'IPCA_mensal': 'PRECOS12_IPCAG12',
    'IPCA_12_meses': 'PRECOS12_IPCAGA12',
    'IPCA_alimentos_bebidas': 'PRECOS12_IPCAAB12',
    'IPCA_artigos_residencia': 'PRECOS12_IPCAAR12',
    'IPCA_comunicacao': 'PRECOS12_IPCACO12',
    'IPCA_habitacao': 'PRECOS12_IPCAHA12',
    'Taxa_desocupacao': 'PNADC12_TDESOC12',
    'BC_exportacoes': 'BPAG12_BCX12',
    'BC_importacoes': 'BPAG12_BCM12',
    'BC_saldo': 'BPAG12_BC12',
    'PIB_industria_geral': 'PIMPFN12_QIIGSNNAS12',
    'PIB_vestuario': 'PIMPFN12_QIIGNN512',
    'PIB_bebidas': 'PIMPFN12_QIIGNN212',
    'PIB_bens_capital': 'PIMPFN12_QIBKNN12',
    'PIB_bens_consumo': 'PIMPFN12_QIBCTNNAS12',
    'PIB_bens_intermediarios': 'PIMPFN12_QIBINNAS12',
    'PIB_celulose_papel': 'PIMPFN12_QIIGNN812',
    'PIB_transformacao': 'PIMPFN12_QIITNN12',
    'PIB_extrativas': 'PIMPFN12_QIEMNNAS12',
    'PIB_insumos_construcao': 'PIMPFN12_YICCNN12',
    'PIB_veiculos': 'PIMPFN12_QIIGNN2012',
    'Exp_agropecuaria': 'FUNCEX12_XVAGP2N12',
    'Imp_agropecuaria': 'FUNCEX12_MVAGP2N12',
    'Imp_veiculos': 'FUNCEX12_MVVEIC2N12',
    'Exp_veiculos': 'FUNCEX_XVVEIC2N',
    'Emplacamento_autoveiculos': 'FENABRAVE12_VENDVETOT12'
}

# Lista para armazenar os DataFrames
tabelas = []

# Loop para baixar e processar todas as s√©ries
for nome, codigo in series.items():
    try:
        df = ipeadatapy.timeseries(codigo)
        df["serie"] = nome  # adiciona coluna com o nome da s√©rie
        tabelas.append(df)
        print(f"S√©rie '{nome}' carregada com sucesso!")
    except Exception as e:
        print(f"Erro ao carregar '{nome}': {e}")

# Junta todas as tabelas em uma √∫nica
df_final = pd.concat(tabelas, ignore_index=True)

# Converte a coluna de data se existir
if 'DATE' in df_final.columns:
    df_final['DATE'] = pd.to_datetime(df_final['DATE'])

# Salva em CSV para importar no Power BI
df_final.to_csv("Dados_Brasil.csv", index=False, encoding="utf-8")

print("\n‚úÖ Todas as s√©ries foram unificadas com sucesso!")
print(f"Total de linhas: {len(df_final)}")
print(f"Colunas: {df_final.columns.tolist()}")
```

# Mexico

```
import pandas as pd
import requests

# --- CONFIGURA√á√ÉO M√âXICO (FONTE: OCDE via DBnomics) ---
# Sem necessidade de Token. Dados est√°veis.
series_mexico = {
    "GDP": "OECD/DSD_NAMAIN10@DF_TABLE1/A.MEX.S1.S1.B1GQ._Z._Z._Z.USD_EXC.V.N.T0101",
    "GDP/Capita": "OECD/DSD_NAMAIN10@DF_TABLE1/A.MEX.S1.S1.B1GQ_POP._Z._Z._Z.USD_EXC_PS.V.N.T0102",
    "Population": "OECD/DSD_POPULATION@DF_POP_HIST/MEX.POP.PS._T._T.H",
    #"Unemployment": "OECD/DSD_LFS@DF_LFS_INDIC/M.MEX.UR._T._T._T.PT_LF.Y"
    "CPI": "OECD/DSD_PRICES@DF_PRICES_ALL/M.MEX.CPI.PA._T.N.IX"

}

def consultar_mexico_oecd(codigo_serie, nome):
    # Usando a API do DBnomics que √© a mais est√°vel para dados da OCDE
    url = f"https://api.db.nomics.world/v22/series/{codigo_serie}?observations=1"

    try:
        response = requests.get(url, timeout=20)

        if response.status_code != 200:
            print(f"‚ùå Erro na fonte para {nome}: C√≥digo {response.status_code}")
            return None

        data = response.json()
        serie = data['series']['docs'][0]

        # Cria DataFrame
        df = pd.DataFrame({
            'Data': serie['period'],
            nome: serie['value']
        })

        # Tratamento
        df['Data'] = pd.to_datetime(df['Data'])
        df.set_index('Data', inplace=True)

        # Pega a data mais recente para valida√ß√£o
        ultima = df.index.max().strftime('%Y-%m-%d')
        print(f"‚úÖ {nome}: Coletado at√© {ultima} ({len(df)} linhas)")
        return df

    except Exception as e:
        print(f"‚ö†Ô∏è Exce√ß√£o em {nome}: {e}")
        return None

# --- EXECU√á√ÉO ---
print("--- Iniciando Coleta M√âXICO (Via OCDE) ---\n")
dfs_mexico = []

for nome, codigo in series_mexico.items():
    df = consultar_mexico_oecd(codigo, nome)
    if df is not None:
        dfs_mexico.append(df)

# --- CONSOLIDA√á√ÉO ---
if dfs_mexico:
    df_final = pd.concat(dfs_mexico, axis=1)
    df_final = df_final.sort_index()

    # C√°lculos Extras
    # 1. Saldo da Balan√ßa
    if "Exportacoes_USD" in df_final.columns and "Importacoes_USD" in df_final.columns:
        df_final["Saldo_Comercial_USD"] = df_final["Exportacoes_USD"] - df_final["Importacoes_USD"]

    # 2. Varia√ß√£o da Infla√ß√£o (O dado vem como √çndice ou Varia√ß√£o dependendo da s√©rie, OCDE costuma mandar √çndice)
    # Se os valores forem > 100, √© √≠ndice. Vamos calcular a varia√ß√£o % s√≥ pra garantir.
    if "Inflacao_Mensal_CPI" in df_final.columns:
         df_final["Inflacao_MoM_Calc"] = df_final["Inflacao_Mensal_CPI"].pct_change() * 100

    arquivo = "Dados_Mexico_Final.csv"
    df_final.to_csv(arquivo, sep=';', decimal=',', encoding='utf-8-sig')

    print(f"\nüéâ Sucesso! Arquivo '{arquivo}' gerado.")
    print(df_final.tail())
else:
    print("\nNenhum dado coletado. Verifique sua conex√£o com a internet.")
```

# Argentina


```
import pandas as pd
import requests
import io

# --- 1. INSIRA SEUS C√ìDIGOS AQUI ---
# Coloque o nome que quiser na esquerda e o SEU C√ìDIGO na direita.
series_ids = {
    "IPC_Mensal": "145.3_INGNACNAL_DICI_M_15",
    "Exporta√ß√£o": "74.3_IET_0_M_16", #Intercambio Comercial Argentino. Valores mensuales
    "Importa√ß√£o": "74.3_IIT_0_M_25",
    "Saldo": "74.3_ISC_0_M_19",
    "IPI": "453.1_SERIE_ORIGNAL_0_0_14_46"
}

def baixar_serie_completa(id_serie, nome):
    # O parametro &limit=5000 garante que venham todas as linhas (n√£o s√≥ as 100 primeiras)
    url = f"https://apis.datos.gob.ar/series/api/series/?ids={id_serie}&limit=5000&format=csv"

    # Headers para evitar bloqueio (Erro 400)
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
    }

    try:
        response = requests.get(url, headers=headers)
        if response.status_code == 200:
            csv_data = io.StringIO(response.text)
            df = pd.read_csv(csv_data)

            # Ajuste b√°sico de colunas
            col_valor = [c for c in df.columns if c != 'indice_tiempo'][0]
            df = df.rename(columns={'indice_tiempo': 'Data', col_valor: nome})
            df['Data'] = pd.to_datetime(df['Data'])
            df.set_index('Data', inplace=True)
            return df
        else:
            print(f"‚ùå Erro ao baixar {nome} (ID: {id_serie}): Status {response.status_code}")
            return None
    except Exception as e:
        print(f"‚ùå Erro cr√≠tico em {nome}: {e}")
        return None

# --- 2. EXECU√á√ÉO ---
dfs = []
print("Iniciando coleta...")

for nome, codigo in series_ids.items():
    if codigo != "INSIRA_OUTRO_CODIGO_AQUI": # Pula o exemplo vazio
        df = baixar_serie_completa(codigo, nome)
        if df is not None:
            dfs.append(df)
            print(f"‚úÖ {nome}: {len(df)} linhas coletadas.")

# --- 3. EXPORTA√á√ÉO ---
if dfs:
    # Junta tudo
    df_final = pd.concat(dfs, axis=1)
    df_final = df_final.sort_index()

    # Salva CSV
    nome_arquivo = "Dados_Argentina.csv"
    df_final.to_csv(nome_arquivo, sep=';', decimal=',', encoding='utf-8-sig')

    print(f"\nArquivo gerado com sucesso: {nome_arquivo}")
else:
    print("Nenhum dado foi coletado.")
```

# Canada


```
import pandas as pd
import pandas_datareader.data as web
import datetime

# 1. Datas
start = datetime.datetime(2010, 1, 1)
end = datetime.datetime.today()

print("--- Iniciando Coleta CANAD√Å (Fonte: FRED) ---")

# 2. Dicion√°rio de C√≥digos (Canad√°)
series_ids = {
    'PIB_Mensal_Nominal': 'NGDPRSAXDCCAQ',   # Exclusividade do Canad√°: PIB Mensal!
    'Desemprego_Taxa': 'LRHUTTTTCAM156S',
    'CPI_Indice': 'CPALTT01CAM659N',           # √çndice base para infla√ß√£o
    'Exportacoes_Bens': 'XTEXVA01CAM667S',
    'Importacoes_Bens': 'XTIMVA01CAM667S',
    'Producao_Industrial': 'CANPROINDMISMEI',
    'Cambio_USD_CAD': 'CCUSSP01CAM650N'        # Taxa de C√¢mbio (Opcional, mas √∫til)
}

try:
    # 3. Coleta via FRED
    df = web.DataReader(list(series_ids.values()), 'fred', start, end)

    # Renomear colunas
    inv_map = {v: k for k, v in series_ids.items()}
    df.columns = [inv_map.get(c, c) for c in df.columns]

    # 4. C√°lculos Autom√°ticos

    # Infla√ß√£o (O FRED entrega o √≠ndice, calculamos a varia√ß√£o %)
    df['Inflacao_MoM'] = df['CPI_Indice'].pct_change(1) * 100
    df['Inflacao_12M'] = df['CPI_Indice'].pct_change(12) * 100

    # Saldo da Balan√ßa Comercial
    if 'Exportacoes_Bens' in df.columns and 'Importacoes_Bens' in df.columns:
        df['Saldo_Balanca'] = df['Exportacoes_Bens'] - df['Importacoes_Bens']

    # 5. Tratamento Final
    df.index = pd.to_datetime(df.index)

    # Exporta√ß√£o
    arquivo = "Dados_Canada.csv"
    df.to_csv(arquivo, sep=';', decimal=',', encoding='utf-8-sig')

    print(f"\n‚úÖ Sucesso! Arquivo gerado: {arquivo}")
    print("√öltimos dados coletados:")
    print(df[['PIB_Mensal_Nominal', 'Inflacao_12M']].tail())

except Exception as e:
    print(f"‚ùå Erro ao coletar dados do Canad√°: {e}")
```

```
import pandas as pd
import requests

# ==============================================================================
# 1. CONFIGURA√á√ÉO DOS PA√çSES
# ==============================================================================
# C√≥digos ISO de 2 letras dos pa√≠ses que voc√™ quer
paises_alvo = {
    "BR": "Brasil",
    "MX": "Mexico",
    "AR": "Argentina",
    "CL": "Chile",
    "CO": "Colombia",
    "PE": "Peru",
    "US": "EUA",
    "CA": "Canada"
}

# Dataset do FMI: International Financial Statistics (IFS)
# Indicador: PCPI_IX (Consumer Price Index, All items, Base 2010=100)
# Estrutura do c√≥digo no DBnomics: IMF/IFS/M.{CODIGO_PAIS}.PCPI_IX
base_url = "https://api.db.nomics.world/v22/series/IMF/IFS"

# ==============================================================================
# 2. MOTOR DE COLETA
# ==============================================================================
print(f"üöÄ Iniciando coleta do FMI (IMF) para {len(paises_alvo)} pa√≠ses...\n")

dfs = []

for iso_code, nome_pais in paises_alvo.items():
    # Monta o ID da s√©rie: M (Mensal) . Pa√≠s . PCPI_IX (√çndice de Pre√ßos)
    series_id = f"M.{iso_code}.PCPI_IX"
    full_url = f"{base_url}/{series_id}?observations=1"

    print(f"üîÑ Buscando {nome_pais} ({series_id})...")

    try:
        response = requests.get(full_url, timeout=20)

        if response.status_code == 200:
            data = response.json()

            # Navega√ß√£o no JSON do DBnomics
            try:
                serie_doc = data['series']['docs'][0]
                datas = serie_doc['period']
                valores = serie_doc['value']

                # Cria DataFrame
                df_temp = pd.DataFrame({
                    'Data': datas,
                    f'CPI_{nome_pais}': valores
                })

                # Tratamento
                df_temp['Data'] = pd.to_datetime(df_temp['Data'])
                df_temp.set_index('Data', inplace=True)

                # Opcional: Calcular infla√ß√£o MoM (Varia√ß√£o Mensal) na hora
                # df_temp[f'Inflacao_MoM_{nome_pais}'] = df_temp[f'CPI_{nome_pais}'].pct_change() * 100

                dfs.append(df_temp)

                # Mostra o √∫ltimo dado para valida√ß√£o
                ultimo = df_temp.index.max().strftime('%Y-%m')
                print(f"   ‚úÖ Sucesso! Dados at√©: {ultimo} ({len(df_temp)} meses)")

            except (IndexError, KeyError):
                print(f"   ‚ö†Ô∏è S√©rie encontrada mas sem dados leg√≠veis.")
        else:
            print(f"   ‚ùå Erro HTTP {response.status_code} (S√©rie n√£o existe no FMI para este pa√≠s?)")

    except Exception as e:
        print(f"   ‚ùå Erro de conex√£o: {e}")

# ==============================================================================
# 3. CONSOLIDA√á√ÉO
# ==============================================================================
if dfs:
    print("\nüîó Unificando tabelas...")

    # Junta tudo pela data
    df_final = pd.concat(dfs, axis=1)
    df_final = df_final.sort_index()

    # Filtra para pegar, por exemplo, de 2010 para c√° (para n√£o ficar um arquivo gigante desde 1950)
    df_final = df_final[df_final.index >= '2010-01-01']

    print("\n--- Resultado Final (√çndices de Pre√ßos - Base 2010=100) ---")
    print(df_final.tail())

    arquivo = "Inflacao_Mensal_IMF_Paises.csv"
    df_final.to_csv(arquivo, sep=';', decimal=',', encoding='utf-8-sig')
    print(f"\nüéâ Arquivo salvo: {arquivo}")
else:
    print("\n‚ùå Nenhum dado coletado.")
```

DB Nomics

```
import requests
import pandas as pd

# ==============================================================================
# 1. CONFIGURA√á√ÉO
# ==============================================================================
paises = {
    "US": "EUA",
    "BR": "Brasil",
    "MX": "M√©xico",
    "CL": "Chile",
    "CO": "Col√¥mbia",
    "PE": "Peru",      # Tende a ser Trimestral no FMI
    "AR": "Argentina"  # Tende a ser Trimestral no FMI
}

# Fonte: FMI (International Financial Statistics)
# C√≥digo: LUR_PT = Unemployment Rate (%)
base_url = "https://api.db.nomics.world/v22/series/IMF/IFS"

print(f"üöÄ Buscando Taxa de Desemprego (% da Popula√ß√£o)...\n")

dfs = []

# ==============================================================================
# 2. MOTOR DE COLETA (Mensal -> Fallback Trimestral)
# ==============================================================================
for iso, nome in paises.items():
    # Tenta Primeiro: MENSAL (M)
    series_id = f"M.{iso}.LUR_PT"
    url = f"{base_url}/{series_id}?observations=1"
    freq_encontrada = "Mensal"

    try:
        response = requests.get(url, timeout=10)
        data = response.json()

        # Verifica se a s√©rie existe e tem dados dentro
        if 'series' not in data or not data['series']['docs'] or not data['series']['docs'][0].get('value'):
             raise ValueError("Vazio")

    except:
        # Tenta Segundo: TRIMESTRAL (Q)
        freq_encontrada = "Trimestral"
        series_id = f"Q.{iso}.LUR_PT"
        url = f"{base_url}/{series_id}?observations=1"

        try:
            response = requests.get(url, timeout=10)
            data = response.json()
            if 'series' not in data or not data['series']['docs']:
                print(f"   ‚ùå {nome}: Dado n√£o encontrado no FMI.")
                continue
        except:
            print(f"   ‚ùå {nome}: Erro de conex√£o.")
            continue

    # --- Extra√ß√£o ---
    serie = data['series']['docs'][0]

    df_temp = pd.DataFrame({
        'Data': serie['period'],
        f'{nome} (%)': serie['value'] # Coluna com nome do pa√≠s
    })

    df_temp['Data'] = pd.to_datetime(df_temp['Data'])
    df_temp.set_index('Data', inplace=True)

    # Arredonda para 1 casa decimal (padr√£o de desemprego)
    df_temp = df_temp.round(2)

    dfs.append(df_temp)
    print(f"   ‚úÖ {nome}: Coletado ({freq_encontrada})")

# ==============================================================================
# 3. CONSOLIDA√á√ÉO
# ==============================================================================
if dfs:
    # Junta tudo pela data
    df_final = pd.concat(dfs, axis=1).sort_index()

    # Filtro: Dados a partir de 2015 para facilitar visualiza√ß√£o
    df_final = df_final[df_final.index >= '2015-01-01']

    print("\n--- Tabela Final: Taxa de Desemprego (%) ---")
    print(df_final.tail(6))

    arquivo = "Desemprego_Taxa_Paises.csv"
    df_final.to_csv(arquivo, sep=';', decimal=',', encoding='utf-8-sig')
    print(f"\nüìÅ Arquivo salvo: {arquivo}")
else:
    print("Nenhum dado encontrado.")
```
